```python
import re
from typing import List, Pattern, Tuple
from urllib.request import urlopen

MOTIF = re.compile(r'(?=(N[^P][ST][^P]))')

def uniport_to_sequence(uniport_id: str) -> Tuple[str, str]:
    '''Given a uniprot ID, returns tuple (id, sequence); 
    attempts with given ID, if 404 error, tries first string 
    in ID ie. x in x_y_z.
    '''
    url = f'http://www.uniprot.org/uniprot/uniprot_id.fasta'
    try:
        responce = urlopen(url.replace('uniprot_id', uniport_id))
    except urllib.error.HTTPError:
        uniport_id = uniport_id.split('_')[0]
        responce = urlopen(url.replace('uniprot_id', uniport_id))
        
    seq = ''.join(responce.read().decode('utf-8').split('\n')[1:])
    
    return uniport_id, seq


def find_motif(
    uniport_id: str,
    sequence: str,
    motif: Pattern
) -> Tuple[str, List[int]]:

    positions = []
    for match in re.finditer(motif, sequence):
        positions.append(match.start() + 1)
    
    return uniport_id, positions

```
