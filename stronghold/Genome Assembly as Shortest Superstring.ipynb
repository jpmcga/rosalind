```python
from collections import defaultdict

def fasta_to_dict(filename):

    file = open(f"{filename}")

    # Get dict from fasta
    sequences = defaultdict(str)

    for line in file:
        line = line.rstrip()
        if line.startswith('>'):
            name = line.strip('>')
        else:
            sequences[name] = sequences[name] + line
    
    return sequences

def find_first_substring(sequences):
    for sequence1 in sequences:
        matches = 0
        prefix = sequence1[:int(len(sequence1)/2)]
        for sequence2 in sequences:
            if sequence1 != sequence2:
                if prefix in sequence2:
                    matches += 1
            if matches == 0:
                return sequence2
        
def best_prefix_match(ref, sequences):
    lst = []
    for sequence in sequences:
        prefix = sequence[:int(len(sequence)/2)]
        lst.append((sequence, ref.rfind(prefix)))
    print(sorted(lst, key = lambda x: x[1], reverse = True)[0][0])
    return sorted(lst, key = lambda x: x[1], reverse = True)[0][0]
```
