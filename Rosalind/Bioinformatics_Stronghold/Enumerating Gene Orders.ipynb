```python
def permutate(target):

    factorial = 1
    for i in range(1, target + 1):
        factorial = factorial * i

    # Initiate an empty list.  This stores each new list
    # and then iterates over it.
    new_list = [[]]

    for x in range(0, target):
        # Takes next number in list
        number = x + 1
        lst_lst = []
        for lst in new_list:
            for pos in range(0, number):
                lst_copy = lst.copy()
                lst_copy.insert(pos, number)
                lst_lst.append(lst_copy)
        new_list = lst_lst.copy()

    return factorial, lst_lst
    
```

```python
res = permutate(5)
print(res[0])
for x in res[1]:
    for i in x:
        print(i, end = " ")
    print()
```

```python
# With permutations
from itertools import permutations

number = 10

# Calculate factorial
factorial = 1
for i in range(1, number + 1):
    factorial = factorial * i

lst = [i for i in range(1, number+1)]

permutations = permutations(lst)

print(factorial)
for permutation in permutations:
    print(permutation)
```

```python
# Lazy answer
import random
# Doesn't work above 8
number = 9

# Calculate factorial
factorial = 1
for i in range(1, number + 1):
    factorial = factorial * i

lst = list(range(1,number+1))
perms = set()

while len(perms) < factorial:
    random.shuffle(lst)
    perms.add(str(lst))

print(factorial)
print(perms)
```
